```python
#Importing the libraries
import pandas as pd
import numpy as np
```

```python
categorical=pd.read_csv("categorical.csv") # Loading the datset
```

```python
categorical.head(10)   # top 10 rows in this case we only have 10 rows in the dataset
```

# ONE HOT ENCODING

```python
df=categorical.drop(["Rec-No"], axis=1)# Dropping the first column
```

```python
df.head(5)
```

```python
# Dummification
```

```python
df=pd.get_dummies(df, prefix=["Temp"], columns=["Temperature"])
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import OneHotEncoder  # to perform encoding
ohc=OneHotEncoder()  #instantiating an object
ohe=ohc.fit_transform(categorical.Temperature.values.reshape(-1,1)).toarray()
dfonehot=pd.DataFrame(ohe, columns=["Temp_"+str(ohc.categories_[0][i]) for i in range (len(ohc.categories_[0]))])
dfh=pd.concat([categorical, dfonehot], axis=1)
dfh.head(3)
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
categorical.loc[:,"Temp_Label_encoder"]=pd.factorize(categorical["Temperature"])[0].reshape(-1,1)
```

```python
categorical.head(5)
```

# Ordinal Encoding

```python
Temp_dict={"cold":1, "warm":2, "hot":3, "very hot":4}
categorical["Temp_ordinal_encoding"]=categorical.Temperature.map(Temp_dict)
```

```python
categorical.head(3)
```

# Binary Encoding

```python
import category_encoders as ce
encoder=ce.BinaryEncoder(cols=["Temperature"])
dfbin=encoder.fit_transform(categorical["Temperature"])
df=pd.concat([categorical, dfbin], axis=1)
df.head(10)
```

# Frequency Encoding

```python
fe=df.groupby("Temperature").size()/len(df)
df.loc[:,"Temp_freq_encoding"]=df["Temperature"].map(fe)
```

```python
df.head()
```

# Mean Encoding

```python
mean_encode=df.groupby("Temperature")["Target"].mean()
print(mean_encode)
```

```python
df.loc[:, "Temperature_mean_enc"]=df["Temperature"].map(mean_encode)
```
